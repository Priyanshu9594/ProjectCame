```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from nltk.corpus import stopwords
from textblob import TextBlob
```

```python
data = pd.read_csv('Messages.csv',encoding="ISO-8859-1")
data.head()
```

```python
#Number of Words 
data['word_count'] = data['1 desk'].apply(lambda x: len(str(x).split(" ")))
data[['1 desk','word_count']].head()
```

```python
#Number of characters 
data['char_count'] = data['1 desk'].str.len() ## this also includes spaces
data[['1 desk','char_count']].head()
```

```python
def avg_word(sentence):
  words = sentence.split()
  return (sum(len(word) for word in words)/len(words))

data['avg_word'] = data['1 desk'].apply(lambda x: avg_word(x))
data[['1 desk','avg_word']].head()
```

```python
freq = pd.Series(' '.join(data['1 desk']).split()).value_counts()
freq.head(60)
```

```python
words = [x.strip() for x in words] # remove both the leading and the trailing characters
words = [x for x in words if words] # removes empty strings, because they are considered in Python as False
words
```
